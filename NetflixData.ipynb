```python
import pandas as pd
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df=df.dropna(subset=["type","title","director","cast","country","date_added","rating"])
```

```python
df.head()
```

```python
df.tail()
```

```python
df["date_added"]=pd.to_datetime(df["date_added"],errors="coerce")
```

```python
df["year_added"]=df["date_added"].dt.year
```

```python
df[["duration_value","duration_uniy"]]=df["duration"].str.extract(r'(\d+)\s*(\w+)')
```

```python
df=df.map(lambda x:x.strip() if isinstance(x,str) else x)
```

```python
print("\nCleaned Data:")
df.head()
```

```python
df.to_csv("netflix_cleaned_data.csv",index=False)
```

```python
print("\nData cleaning complete ! File saved as 'netflix_cleaned_data.csv' ")
```
